```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mean = [0, 0]
covariance = [[1, 0], [0, 100]]
```

```python
ds = np.random.multivariate_normal(mean, covariance, 500)
```

```python
plt.plot(ds[:, 0])
```

```python
plt.plot(ds[:, 1])
```

```python
ds[:, 0].mean()
```

```python
ds[:, 0].var()
```
