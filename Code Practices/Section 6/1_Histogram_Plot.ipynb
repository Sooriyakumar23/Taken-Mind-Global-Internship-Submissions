```python
# Importing Libraries

from numpy.random import randn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data creation

ds1 = randn(100)
ds2 = randn(1000)
ds3 = randn(10000)
```

```python
# Mean of each dataset

ds1.mean(), ds2.mean(), ds3.mean()
```

```python
# Variance of each dataset

ds1.var(), ds2.var(), ds3.var()
```

## Matplotlib Histogram plots

#### ds1

```python
plt.hist(ds1)
plt.savefig('matplotlib_ds1.png')
```

#### ds2 

```python
plt.hist(ds2)
plt.savefig('matplotlib_ds2.png')
```

#### ds3

```python
plt.hist(ds3)
plt.savefig('matplotlib_ds3.png')
```

#### ds2, ds3

```python
plt.hist(ds1, density=True, color='green', bins=15, alpha=0.5)
plt.hist(ds3, density=True, bins=15, alpha=0.5)
plt.savefig('matplotlib_ds1_ds3.png')
```

## Seaborn Histogram plots 

```python
# Data creation

ds3 = randn(1000)
ds4 = randn(1000)
```

```python
sns_scatter = sns.jointplot(ds3, ds4)
sns_scatter.savefig('scatter_ds3_ds4.png')
```

```python
sns_hex = sns.jointplot(ds3, ds4, kind='hex')
sns_hex.savefig('hex_ds3_ds4.png')
```
