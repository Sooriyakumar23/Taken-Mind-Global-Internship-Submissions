## KDE - Kernel Density Estimation 

<b> $*$ A non-parametric way to estimate the probability density function of a random variable  </b>

Kernel (K) :- Non negative integer value

* Famous concept in image processing/deep learning(convolution), linux(intermediate between hardware & shell/user), machine learning(SVM), statistics


 h :- Smoothing function / bandwidth

$K_h$ :- Scalled kernel

<a href="https://en.wikipedia.org/wiki/Kernel_density_estimation"> KDE - Wikipedia </a>

```python
from numpy.random import randn
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
ds = randn(100)
```

```python
sns.rugplot(ds)
```

```python
plt.hist(ds, alpha=1.0)
```

```python
x_axes = np.linspace(ds.min()-1, ds.max()+1, 50)
```

```python
plt.plot(x_axes)
```

Bandwidth : <img src="https://wikimedia.org/api/rest_v1/media/math/render/svg/b1538fbae068efc2dc9ca34513760df51e492c28">

```python
h = ((4*(ds.std())**5)/(3*len(ds)))**(1/5)
```

```python
h
```

```python
kernels = []
```

```python
for point in ds:
    kernel = stats.norm(point, h).pdf(x_axes)
    kernels.append(kernel)
    
    kernel /=  kernel.max()
    kernel *= 0.6
    plt.plot(x_axes, kernel, color='red', alpha=0.6)
    
plt.ylim(0, 1)
plt.hist(ds)
```

```python
kde = np.sum(kernels, axis=0)
```

```python
plt.plot(x_axes, kde, color='green', alpha=0.6)
sns.rugplot(ds)
```
